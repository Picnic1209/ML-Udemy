There are two ways to plot data in matplotlib
first is pyplot API and the second is object-oriented API

### object-oriented is more advanced and we should try to use that

```python
# 1st method
fig = plt.figure() #creates a figure, just the box
ax = fig.add_subplot() #adds some axes, or in other words, ax is the name of a graph
plt.show()
```

```python
# 2nd method
fig = plt.figure() #creates a figure
ax = fig.add_axes([2,5,1,1])
ax.plot(x,y) #add some data
plt.show()
```

```python
# 3rd method (recommended)
fig,ax = plt.subplots()
ax.plot(x,y); #add some data
type(fig), type(ax)
```

# Matplotlib Example workflow

```python
#Step 0. Import matplotlib and get ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

#1. Prepare data
x= [1,2,3,4]
y=[11,22,33,44]

#2. Setup plot
fig, ax= plt.subplots(figsize=(5,10))

#3. Plot Data
ax.plot(x,y)

#4. Customize plot
ax.set(title="simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

#5. Save & show(you save the whole figure)
fig.savefig("./img/sample-plot.png")
```

## Two options for subplots

```python
# subplots option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                        ncols=2,
                                        figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10))
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# doubt: how to display only a single plot??
#plot to eac different axis
fig,ax=plt.subplots()
ax.plot(x,x/2);

```

```python
#subplot option 2
fig,ax = plt.subplots(nrows=2,
                      ncols=2,
                     figsize=(10,5))
# plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(100));
```

### this plot isnt ideal

### Which one should we use?? (pyplot vs matplotlib OO method?)

* When plotting something quick, just use pyplot method
* When plotting something advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
len(over_50)
```

```python
 #pyplot method
over_50.plot(kind="scatter",
             x= 'age',
             y='chol',
             c='target',
             figsize=(10,6));
```

```python
# OO method mixed with pyplot method
fig, ax= plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x='age',
            y='chol',
            c='target',
            ax=ax);
# c means color
# ax.set_xlim([45,100]);
```

```python
## OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"] )
# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
      ylabel="Cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(), title ="Targetw");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex= True);

#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y = over_50["chol"],
                      c=over_50["target"])
# customize ax0
ax0.set(title="Heart disease and cholestrol level",
        ylabel="Cholestrol")
# Add a legend
ax0.legend(*scatter.legend_elements(),title="Target")
# Add meanline
ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--');

##add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
#customize ax1
ax1.set(title="Heart disease and max heart rate",
        xlabel="Age",
        ylabel="highest heart rate")

#add a legend
ax1.legend(*scatter1.legend_elements(),title="Targets")

#addmeanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle='--');

## add a title to the entire figure
fig.suptitle("Heart Disease Analysis",fontsize=40,fontweight='bold');
```

```python
#set the style
plt.style.use('seaborn-whitegrid')

## OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

#plot the data
scatter = ax.scatter(x = over_50["age"],
                     y = over_50["chol"],
                     c = over_50["target"],
                     cmap='winter') # it changes colour scheme
# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
      ylabel="Cholestrol");

# add a legend
ax.legend(*scatter.legend_elements(), title ="Targetw");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# Customizing x and y axis limitation

# subplot of chol, age, thalach

fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex= True);

#add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y = over_50["chol"],
                      c=over_50["target"],
                      cmap="winter")
# customize ax0
ax0.set(title="Heart disease and cholestrol level",
        ylabel="Cholestrol")

#set x limit
ax0.set_xlim([50,80])

# Add a legend
ax0.legend(*scatter.legend_elements(),title="Target")
# Add meanline

ax0.axhline(y=over_50['chol'].mean(),
           linestyle='--');

##add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                       cmap="winter")
#customize ax1
ax1.set(title="Heart disease and max heart rate",
        xlabel="Age",
        ylabel="highest heart rate")

#change ax1 x and y axis limit
ax1.set_xlim([50,80])
ax1.set_ylim([50,200])

#add a legend
ax1.legend(*scatter1.legend_elements(),title="Targets")

#addmeanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle='--');

## add a title to the entire figure
fig.suptitle("Heart Disease Analysis",fontsize=40,fontweight='bold');
```
